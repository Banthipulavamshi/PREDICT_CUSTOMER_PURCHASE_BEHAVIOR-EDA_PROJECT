```python
sns.countplot(x="PurchaseStatus", data=data)

plt.show()
```

```python
sns.scatterplot(x="Age", y="NumberOfPurchases", data=data)
plt.xlabel("Age")
plt.ylabel("Number of Purchases")
plt.title("Age vs Number of Purchases")
plt.show()
```

```python
plt.scatter(data["NumberOfPurchases"], data["TimeSpentOnWebsite"])
plt.xlabel("Number Of Purchases")  
plt.ylabel("Time Spent On Website")
plt.title("Number Of Purchases Price vs. Time Spent On Website")
plt.show()
```

```python
plt.scatter(data["AnnualIncome"], data["TimeSpentOnWebsite"])
plt.xlabel("Annual Income")
plt.ylabel("Time Spent On Website")
plt.title("Annual Income vs Time Spent On Website")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(data['AnnualIncome'], bins=20, edgecolor='black')
plt.xlabel('Annual Income')
plt.ylabel('Frequency')
plt.title('Distribution of Annual Income')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(data['AnnualIncome'], data['NumberOfPurchases'], alpha=0.5)
plt.xlabel('Annual Income')
plt.ylabel('Number of Purchases')
plt.title('Number of Purchases vs Annual Income')
plt.show()
```

```python
avg_purchases = data.groupby('ProductCategory')['NumberOfPurchases'].mean().sort_values()
plt.figure(figsize=(12, 8))
avg_purchases.plot(kind='bar', color='orange')
plt.xlabel('Product Category')
plt.ylabel('Average Number of Purchases')
plt.title('Average Number of Purchases by Product Category')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(15, 9))
data.boxplot(column='TimeSpentOnWebsite', by='LoyaltyProgram', grid=False)
plt.xlabel('Loyalty Program Status')
plt.ylabel('Time Spent on Website')
plt.title('Time Spent on Website by Loyalty Program Status')
plt.suptitle('')  # Suppress the default title
plt.show()
```

```python
status_counts = data['PurchaseStatus'].value_counts()
plt.figure(figsize=(5, 5))
plt.pie(status_counts, labels=status_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Purchase Status')
plt.show()
```

```python
plt.figure(figsize=(15, 9))
plt.hist(data['DiscountsAvailed'], bins=20, edgecolor='black')
plt.xlabel('Discounts Availed')
plt.ylabel('Frequency')
plt.title('Distribution of Discounts Availed')
plt.show()
```

```python
sns.scatterplot(x="Age", y="NumberOfPurchases", data=data)

plt.xlabel("Age")
plt.ylabel("Number of Purchases")
plt.title("Number of Purchases by Age")
plt.show()
```

```python
sns.boxplot(x="Gender", y="NumberOfPurchases", data=data)

plt.xlabel("Gender")
plt.ylabel("Number of Purchases")
plt.title("Number of Purchases by Gender")
plt.show()
```

```python
sns.violinplot(x="ProductCategory", y="NumberOfPurchases", data=data)

plt.xlabel("Product Category")
plt.ylabel("Number of Purchases")
plt.title("Purchase Distribution by Product Category")
plt.show()
```

```python
corr_matrix = data.select_dtypes(include=['number']).corr()

# Create a heatmap
plt.figure(figsize=(15, 15))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", linewidths=0.5)

plt.title("Correlation Heatmap")
plt.show()
```
